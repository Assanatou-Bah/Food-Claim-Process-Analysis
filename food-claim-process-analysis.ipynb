```python
# Importng necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data

food_claims = pd.read_excel(r"C:\Users\ams_j\Desktop\Case_Study\food_claims_.xlsx", sheet_name = 'food_claims')
```

```python
# Viewing the data

food_claims 
```

```python
# Checking info of the data

food_claims.info()
```

```python
# Checking missing values

food_claims.isna().sum()
```

```python
# Number of claims for each location

food_claims['location'].value_counts()
```

```python
# Creating a plot for number of claims in each location

sns.countplot(x='location', data=food_claims, palette='Blues')
plt.title('Number of Claims Per Location')


# Display the plot
plt.show()
```

```python
# Creating a plot for the distribution of time to close for all claims

sns.set_style("darkgrid")

sns.histplot(data=food_claims, x = 'time_to_close', bins = 40, color="#6096ba")
plt.title('Range in time to close')


# Display the plot
plt.show()
```

```python
# Creatng a plot for relationship between time to close and location

sns.boxplot(data= food_claims, x='location', y='time_to_close', palette='Blues', width=0.5)


# Display the plot
plt.show()
```

```python
#subset for only time to close between 90 and 300 to remove outliers

food_claims_clean = food_claims.drop(food_claims[(food_claims['time_to_close'] > 90) & (food_claims['time_to_close'] <= 300)].index)

# Creatng a plot for relationship between time to close and location without outliers

sns.boxplot(data= food_claims_clean, x='location', y='time_to_close', palette='Blues')


# Display the plot
plt.show()
```
